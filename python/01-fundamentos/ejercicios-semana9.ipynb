```python
def resumen_ventas(ventas):
  total = 0
  precio_mas_alto = 0
  producto_mas_caro = ""

  for v in ventas:
    total = total + (v["precio"] * v["cantidad"])

    if v["precio"] > precio_mas_alto :
      precio_mas_alto = v["precio"]
      producto_mas_caro = v["producto"]
  return total, producto_mas_caro

total, producto = resumen_ventas(ventas)
print(f"Total ingresos: ${total}")
print(f"Producto mas caro: {producto}")


```
